7. Постройте круговую диаграмму отображающую выживаемость мужчин и женщин в разных классах обслуживания(визуализация должна быть интуитивно понятная без объяснений, женщины и мужчины соответствующих классов объеденены визуально в одну группу и 2 подгруппы)*

\* Задание на доп. балл

````python
data = df.groupby(['Sex', 'Pclass', 'Survived']).size().unstack().fillna(0)
data.columns = ['Не выжили', 'Выжили']

male_data = data.loc['male']
female_data = data.loc['female']

labels = ['Класс 1 - Не выжили', 'Класс 1 - Выжили', 
          'Класс 2 - Не выжили', 'Класс 2 - Выжили', 
          'Класс 3 - Не выжили', 'Класс 3 - Выжили']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 8), dpi=300)

ax1.pie(male_data.values.flatten(), labels=labels, autopct='%1.1f%%', startangle=140)
ax1.set_title('Выживаемость Мужчин')

ax2.pie(female_data.values.flatten(), labels=labels, autopct='%1.1f%%', startangle=140)
ax2.set_title('Выживаемость Женщин')

plt.suptitle('Выживаемость по Полу и Классу Обслуживания')

plt.subplots_adjust(left=0.1, right=0.9, wspace=0.5)
plt.show()
````
